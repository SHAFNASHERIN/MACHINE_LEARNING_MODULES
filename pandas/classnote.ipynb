```python
#pandas merging,joinng and concatenating
df_customer = pd.DataFrame({
    'id':[1,2,3,4],
    'name':['Tom','Jenny','James','Dan']
})
df_info = pd.DataFrame({
    'id':[2,3,4,5],
    'age':[31,20,40,70],
    'sex':['F','M','M','F']
})
```

```python
#merging
pd.merge(df_customer,df_info)
```

```python
df_customer.merge(df_info)
```

```python
pd.merge(df_customer,df_info,on='id')
```

```python
# To combine data on multiple common columns, you can pass a list toon:

# with multiple key columns
df_order = pd.DataFrame({
    'id': [2,3,4,5],
    'name': ['Jenny', 'James', 'Dan', 'leo'],
    'quantity': [2,4,6,10]
})

pd.merge(df_customer, df_order, on=['id', 'name'])
```

```python
df_customer = pd.DataFrame({
    'id':[1,2,3,4],
    'name':['Tom','Jenny','James','Dan']
})
df_info = pd.DataFrame({
    'id':[2,3,4,5],
    'age':[31,20,40,70],
    'sex':['F','M','M','F']
})
```

```python
df_info_2 = pd.DataFrame({
    'customer_id': [2,3,4,5],
    'age': [31,20,40,70],
    'sex': ['F', 'M', 'M', 'F']
})

pd.merge(
  df_customer, 
  df_info_2, 
  left_on='id', 
  right_on='customer_id'
)
```

```python
df_customer = pd.DataFrame({
    'id': [1,2,3,4],
    'name': ['Tom', 'Jenny', 'James', 'Dan'],
})
df_info = pd.DataFrame({
    'id': [2,3,4,5],
    'age': [31,20,40,70],
    'sex': ['F', 'M', 'M', 'F']
})
pd.merge(df_customer, df_info, on='id')
```

```python
pd.merge(df_customer, df_info, on='id')
```

```python
# To explicitly specify the inner join, you can set the argument how='inner'

pd.merge(df_customer, df_info, how='inner', on='id')
```

```python
#left join
pd.merge(df_customer, df_info, how='left', on='id')
```

```python
#right join
pd.merge(df_customer, df_info, how='right', on='id')
```

```python
#outer join
pd.merge(df_customer, df_info, how='outer', on='id')
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df2
```

```python
pd.merge(df1,df2,on='city',how='outer',indicator=True)
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```
