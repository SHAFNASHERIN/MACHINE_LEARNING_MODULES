```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Day=[1,2,3,4,5,6,7]
Temp=[50,51,52,48,47,49,46]
```

```python
plt.plot(Day,Temp)
plt.show()
```

```python
plt.plot(Day,Temp,color='green',linestyle='dashdot',marker='o',linewidth=2)
plt.show()
```

```python
plt.xlabel('Day',fontsize=10)
plt.ylabel('Temperature',fontsize=10)
plt.title("weather_info")
plt.plot(Day,Temp,color='green',linestyle='dashdot',marker='o',linewidth=2,alpha=0.4)
plt.show()
```

```python
#legend,Grid
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel('Day',fontsize=10)
plt.ylabel('Temperature',fontsize=10)
plt.title("weather_info")

plt.plot(days,max_t,label="max")
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label="avg")
plt.legend(shadow=True,fontsize="large")
plt.grid()
plt.show()
```

```python
#Bar chart
# Simple bar chart showing revenues of major US tech companies

company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
plt.bar(company,revenue)
plt.show()
```

```python
plt.xlabel("Major companies")
plt.ylabel("Revenue")
plt.title("Revenue Model for Top Companies")
plt.bar(company,revenue)
plt.show()
```

```python
plt.bar(company,revenue,label="Revenue")
plt.bar(company,profit,label="Profit")
plt.legend()
plt.show()
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(xpos,revenue)
```

```python
plt.bar(xpos+2,revenue)
```

```python
plt.xticks(xpos,company)
plt.xlabel("Company Name")
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.bar(xpos-0.2,revenue,width=0.4,label="Revenue")
plt.bar(xpos+0.2,profit,width=0.8,label="profit")
plt.legend()
plt.show()
```

```python
plt.barh(xpos,revenue)
```

```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np


# Creating dataset
np.random.seed(10)

data_1 = np.random.normal(100, 10, 200)
data_2 = np.random.normal(90, 20, 200)
data_3 = np.random.normal(80, 30, 200)
data_4 = np.random.normal(70, 40, 200)
data = [data_1, data_2, data_3, data_4]

fig = plt.figure(figsize =(10, 7))


# Creating plot
plt.boxplot(data,vert=True,patch_artist=False)

# show plot
plt.show()
```

```python
plt.hist(data,rwidth=0)
```

```python
plt.figure(figsize=(10,5))
plt.show()
```
